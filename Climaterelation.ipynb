```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp_df = pd.read_csv("GlobalTemperatures.csv")
temp_df['dt'] = pd.to_datetime(temp_df['dt'])
temp_df['Year'] = temp_df['dt'].dt.year

temp_df = temp_df[['Year', 'LandAverageTemperature']]

temp_df = temp_df.dropna()
yearly_temp = temp_df.groupby('Year')['LandAverageTemperature'].mean()
```

```python
rain_df = pd.read_csv("Rainfall_1901_2016_PAK.csv")
rain_df.columns = rain_df.columns.str.strip()
rain_df = rain_df[['Year', 'Rainfall - (MM)']]
rain_df = rain_df.dropna()
yearly_rain = rain_df.groupby('Year')['Rainfall - (MM)'].sum()
```

```python
yearly_temp = yearly_temp.reset_index()

yearly_rain = yearly_rain.reset_index()
```

```python
climate_df = pd.merge(
    yearly_temp,
    yearly_rain,
    on='Year'
)

```

```python
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")



sns.scatterplot(
    data=climate_df,
    x='LandAverageTemperature',
    y='Rainfall - (MM)'
)

plt.title(
    "Temperature vs Rainfall Relationship",
    fontsize=16,
    fontweight='bold'
)
plt.xlabel("Average Temperature (°C)")
plt.ylabel("Rainfall (MM)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.set_style("whitegrid")

sns.regplot(
    data=climate_df,
    x='LandAverageTemperature',
    y='Rainfall - (MM)',
    scatter_kws={'alpha':0.7},
    line_kws={'color':'red'}
)
plt.title(
    "Temperature vs Rainfall Relationship",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Average Temperature (°C)")
plt.ylabel("Rainfall (MM)")

plt.tight_layout()

plt.show()
```

“This analysis explores whether long-term global temperature changes show any visible relationship with rainfall patterns in my country Pakistan. A scatter plot with regression trend line is used to identify possible correlations between the two climate variables.”
